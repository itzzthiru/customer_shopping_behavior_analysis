```python
from tabulate import tabulate
#Simple Function to run and print queries
def run_query(sql_query):
    try:
        with engine.connect() as conn:
            result_df = pd.read_sql(sql_query, conn)
            print(tabulate(result_df, headers='keys', tablefmt='grid', showindex=False))
    except Exception as e:
        print(f"Error executing query: {e}")
```

```python
# Execute your queries line by line 
print("Query 1: Top 5 Rows")
run_query("SELECT * FROM customer LIMIT 5;")
```

```python
print("Query 2: What is the total revenue generated by male vs. female customers?")
run_query("""SELECT gender, SUM(purshase_amount) AS revenue 
FROM customer 
GROUP BY gender""")
```

```python
print("Query 3: Which customers used a discount but still spent more than the average purchase amount? ")
run_query("""SELECT customer_id, purshase_amount  
          FROM customer WHERE discount_applied    ='Yes' 
          AND purshase_amount  >= (SELECT AVG(purshase_amount ) FROM customer)""")
```

```python
print("Query 4: Which are the top 5 products with the highest average review rating?")
run_query("""SELECT item_purchased, ROUND(AVG(review_rating), 2) AS "Average Product Rating"
FROM customer
GROUP BY item_purchased
ORDER BY AVG(review_rating) DESC
LIMIT 5""")
```

```python
print("Query 5: Compare the average Purchase Amounts between Standard and Express Shipping. ")
run_query("""SELECT shipping_type, 
       ROUND(AVG(purshase_amount), 2) AS avg_purchase
FROM customer
WHERE shipping_type IN ('Standard', 'Express')
GROUP BY shipping_type""")
```

```python
print("Query 6: Do subscribed customers spend more? Compare average spend and total revenue ")
run_query("""SELECT subscription_status,
       COUNT(customer_id) AS total_customers,
       ROUND(AVG(purshase_amount), 2) AS avg_spend,
       ROUND(SUM(purshase_amount), 2) AS total_revenue
FROM customer
GROUP BY subscription_status
ORDER BY total_revenue DESC, avg_spend DESC""")
```

```python
print("Query 7: Which 5 products have the highest percentage of purchases with discounts applied?")
run_query("""SELECT item_purchased,
       ROUND(100.0 * SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS discount_rate
FROM customer
GROUP BY item_purchased
ORDER BY discount_rate DESC
LIMIT 5""")
```

```python
print("""Query 8: Segment customers into New, Returning, and Loyal based on their total 
number of previous purchases, and show the count of each segment.""")
run_query("""WITH customer_type AS (
    SELECT customer_id, previous_purchases,
    CASE 
        WHEN previous_purchases = 1 THEN 'New'
        WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
        ELSE 'Loyal'
    END AS customer_segment
    FROM customer
)
SELECT customer_segment, COUNT(*) AS "Number of Customers" 
FROM customer_type 
GROUP BY customer_segment""")
```

```python
print("Query 9: What are the top 3 most purchased products within each category? ")
run_query("""WITH item_counts AS (
    SELECT category,
           item_purchased,
           COUNT(customer_id) AS total_orders,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(customer_id) DESC) AS item_rank
    FROM customer
    GROUP BY category, item_purchased
)
SELECT item_rank, category, item_purchased, total_orders
FROM item_counts
WHERE item_rank <= 3""")
```

```python
print("""Query 10: Are customers who are repeat buyers
      (more than 5 previous purchases) also likely to subscribe?""")
run_query("""SELECT subscription_status,
       COUNT(customer_id) AS repeat_buyers
FROM customer
WHERE previous_purchases > 5
GROUP BY subscription_status""")
```

```python
print("Query 11: What is the revenue contribution of each age group?")
run_query("""SELECT 
    age_group,
    SUM(purshase_amount) AS total_revenue
FROM customer
GROUP BY age_group
ORDER BY total_revenue DESC""")
```
